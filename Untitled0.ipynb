```python
import gradio as gr
from PIL import Image, ImageDraw, ImageFont

def annotate(image: Image.Image, text: str, shape: str):
    print(f"Input image type: {type(image)}")
    if image is None:
        print("Input image is None")
        return None

    # Create a drawing object
    draw = ImageDraw.Draw(image)

    # Add text to the image
    try:
        font = ImageFont.truetype("arial.ttf", 40)
    except IOError:
        font = ImageFont.load_default()
        print("Arial font not found, using default PIL font.")

    text_color = (255, 0, 0) # Red color
    text_position = (10, 10) # Top-left corner

    if text:
        draw.text(text_position, text, fill=text_color, font=font)

    # Add shape to the image
    shape_color = (0, 0, 255) # Blue color
    shape_position = [(50, 50), (150, 150)] # Example position for a rectangle

    if shape == "rectangle":
        draw.rectangle(shape_position, outline=shape_color, width=5)
    elif shape == "ellipse":
        draw.ellipse(shape_position, outline=shape_color, width=5)
    # Add more shapes here as needed

    print(f"Output image type: {type(image)}")
    return image

with gr.Blocks() as demo:
    gr.Markdown("## üñäÔ∏è Image Annotation Tool\nUpload an image, draw freely, then click Submit.")
    with gr.Row():
        inp = gr.Image(label="Upload & Draw", type="pil", interactive=True, image_mode="RGBA")
        out = gr.Image(label="Annotated Result")
    with gr.Row():
        text_input = gr.Textbox(label="Text to add")
        shape_dropdown = gr.Dropdown(label="Shape to add", choices=["None", "rectangle", "ellipse"], value="None")
    btn = gr.Button("Submit")
    btn.click(annotate, inputs=[inp, text_input, shape_dropdown], outputs=out)

demo.launch(share=True)
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLO model
model = YOLO('yolov8n.pt')
```

## Perform Object Detection

### Subtask:
Load an image and use the pre-trained model to detect objects, specifically humans.

**Reasoning**:
Load a sample image using OpenCV and then use the `model.predict()` method to perform object detection on the image.

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

# Load a sample image
# You can replace 'path/to/your/image.jpg' with the actual path to your image file
# For demonstration, let's download a sample image
# !curl -o image.jpg https://upload.wikimedia.org/wikipedia/commons/thumb/e/e4/Cars_in_traffic_at_night.jpg/1280px-Cars_in_traffic_at_night.jpg

# To upload an image manually, use the following code:
uploaded = files.upload()

for filename in uploaded.keys():
  image_path = filename
  image = cv2.imread(image_path)

# Perform object detection
results = model.predict(image)

# Display the image with detections (optional)
# results[0].plot() can be used to plot bounding boxes on the image
# For displaying in Colab, we can use cv2_imshow or matplotlib
annotated_image = results[0].plot()
cv2_imshow(annotated_image)
```

## Count the Number of Humans

### Subtask:
Process the object detection results to count the number of instances of the "person" class.

**Reasoning**:
Iterate through the detection results and count the bounding boxes that correspond to the 'person' class.

```python
# Count the number of humans
human_count = 0
for r in results:
    # Assuming 'person' is class 0 in COCO dataset, which YOLOv8 uses
    # You might need to adjust the class index based on the model used
    person_class_id = 0
    human_count = len([box for box in r.boxes if box.cls == person_class_id])

print(f"Number of humans detected: {human_count}")
```

## Count Other Detected Objects

### Subtask:
Process the object detection results to count the number of instances of other detected object classes besides "person".

**Reasoning**:
Iterate through the detection results and count the bounding boxes that correspond to each detected class, excluding the 'person' class.

```python
from collections import Counter

# Get the class names from the model
class_names = model.names

# Collect all detected class IDs (excluding 'person')
other_objects_class_ids = [
    int(box.cls) for r in results for box in r.boxes if int(box.cls) != person_class_id
]

# Count the occurrences of each class ID
object_counts = Counter(other_objects_class_ids)

# Display the counts of other objects
print("\nCounts of other detected objects:")
if object_counts:
    for class_id, count in object_counts.items():
        class_name = class_names[class_id]
        print(f"{class_name}: {count}")
else:
    print("No other objects detected besides humans.")
```

## Add confidence filtering

### Subtask:
Introduce a step to filter the detection results based on a user-defined confidence threshold.


**Reasoning**:
Define the confidence threshold and filter the detection results for both human and other object counts based on this threshold, then print the filtered counts.



```python
# Define a confidence threshold
confidence_threshold = 0.5

# Count the number of humans with confidence above the threshold
human_count_filtered = 0
for r in results:
    person_class_id = 0
    human_count_filtered = len([box for box in r.boxes if int(box.cls) == person_class_id and box.conf >= confidence_threshold])

print(f"Number of humans detected (confidence >= {confidence_threshold}): {human_count_filtered}")

# Count other detected objects with confidence above the threshold
other_objects_class_ids_filtered = [
    int(box.cls) for r in results for box in r.boxes if int(box.cls) != person_class_id and box.conf >= confidence_threshold
]

object_counts_filtered = Counter(other_objects_class_ids_filtered)

# Display the filtered counts of other objects
print(f"\nCounts of other detected objects (confidence >= {confidence_threshold}):")
if object_counts_filtered:
    for class_id, count in object_counts_filtered.items():
        class_name = class_names[class_id]
        print(f"{class_name}: {count}")
else:
    print("No other objects detected besides humans with the specified confidence threshold.")
```

## Calculate bounding box area

### Subtask:
Add code to calculate and potentially display the area of the bounding boxes for detected objects.


**Reasoning**:
Iterate through the detection results, calculate the area of each bounding box, and print the area along with the class name and confidence score.



```python
# Calculate and display the area of bounding boxes
print("\nDetected object areas:")
for r in results:
    for box in r.boxes:
        # Get bounding box coordinates in xyxy format
        x1, y1, x2, y2 = box.xyxy[0].tolist()

        # Calculate area
        area = (x2 - x1) * (y2 - y1)

        # Get class name and confidence
        class_id = int(box.cls)
        class_name = class_names[class_id]
        confidence = box.conf.item()

        print(f"  {class_name} (Confidence: {confidence:.2f}): Area = {area:.2f}")
```

## Integrate with gradio

### Subtask:
Modify the existing Gradio interface (or create a new one) to include the object detection and counting functionality.


**Reasoning**:
Define the `process_image_for_gradio` function to handle image processing and object detection, then set up the Gradio interface with input/output components and link the button click event to the function.



```python
import cv2
import numpy as np
from PIL import Image

def process_image_for_gradio(image: Image.Image):
    """
    Processes a PIL image to perform object detection and count humans and other objects.

    Args:
        image: A PIL Image object.

    Returns:
        A tuple containing:
            - Annotated PIL Image with bounding boxes, labels, and confidence.
            - Integer count of humans detected.
            - Dictionary of counts for other detected objects.
    """
    if image is None:
        return None, 0, {}

    # Convert PIL Image to OpenCV format (BGR)
    image_cv2 = cv2.cvtColor(np.array(image), cv2.COLOR_RGBA2BGR)

    # Perform object detection
    results = model.predict(image_cv2)

    # Count humans and other objects
    human_count = 0
    other_objects_class_ids = []
    person_class_id = 0 # Assuming 'person' is class 0

    for r in results:
        for box in r.boxes:
            class_id = int(box.cls)
            if class_id == person_class_id:
                human_count += 1
            else:
                other_objects_class_ids.append(class_id)

    object_counts = Counter(other_objects_class_ids)
    class_names = model.names
    other_object_counts_named = {class_names[class_id]: count for class_id, count in object_counts.items()}


    # Annotate the image
    annotated_image_cv2 = results[0].plot(conf=True)

    # Convert annotated OpenCV image back to PIL format (RGB)
    annotated_image_pil = Image.fromarray(cv2.cvtColor(annotated_image_cv2, cv2.COLOR_BGR2RGB))


    return annotated_image_pil, human_count, other_object_counts_named

# Modify the existing Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## üö∂‚Äç‚ôÇÔ∏è Object Detection and Counting")
    with gr.Row():
        inp_image = gr.Image(label="Upload Image", type="pil", interactive=True)
        out_image = gr.Image(label="Annotated Image")
    with gr.Row():
        human_count_output = gr.Number(label="Number of Humans Detected")
        other_objects_output = gr.Textbox(label="Other Objects Detected")
    btn = gr.Button("Detect Objects")

    btn.click(
        process_image_for_gradio,
        inputs=inp_image,
        outputs=[out_image, human_count_output, other_objects_output]
    )

demo.launch(share=True)
```

## Save annotated image

### Subtask:
Add code to save the image with the detection results drawn on it.


**Reasoning**:
Modify the `process_image_for_gradio` function to save the annotated image to a file with a unique filename using the current timestamp.



```python
import cv2
import numpy as np
from PIL import Image
from collections import Counter
import time
import os

def process_image_for_gradio(image: Image.Image):
    """
    Processes a PIL image to perform object detection and count humans and other objects,
    and saves the annotated image.

    Args:
        image: A PIL Image object.

    Returns:
        A tuple containing:
            - Annotated PIL Image with bounding boxes, labels, and confidence.
            - Integer count of humans detected.
            - Dictionary of counts for other detected objects.
    """
    if image is None:
        return None, 0, {}

    # Convert PIL Image to OpenCV format (BGR)
    image_cv2 = cv2.cvtColor(np.array(image), cv2.COLOR_RGBA2BGR)

    # Perform object detection
    results = model.predict(image_cv2)

    # Count humans and other objects
    human_count = 0
    other_objects_class_ids = []
    person_class_id = 0 # Assuming 'person' is class 0

    for r in results:
        for box in r.boxes:
            class_id = int(box.cls)
            if class_id == person_class_id:
                human_count += 1
            else:
                other_objects_class_ids.append(class_id)

    object_counts = Counter(other_objects_class_ids)
    class_names = model.names
    other_object_counts_named = {class_names[class_id]: count for class_id, count in object_counts.items()}


    # Annotate the image
    annotated_image_cv2 = results[0].plot(conf=True)

    # Save the annotated image
    timestamp = int(time.time())
    save_dir = "annotated_images"
    os.makedirs(save_dir, exist_ok=True)
    save_path = os.path.join(save_dir, f"annotated_image_{timestamp}.png")
    cv2.imwrite(save_path, annotated_image_cv2)
    print(f"Annotated image saved to: {save_path}")


    # Convert annotated OpenCV image back to PIL format (RGB)
    annotated_image_pil = Image.fromarray(cv2.cvtColor(annotated_image_cv2, cv2.COLOR_BGR2RGB))


    return annotated_image_pil, human_count, other_object_counts_named

# Modify the existing Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## üö∂‚Äç‚ôÇÔ∏è Object Detection and Counting")
    with gr.Row():
        inp_image = gr.Image(label="Upload Image", type="pil", interactive=True)
        out_image = gr.Image(label="Annotated Image")
    with gr.Row():
        human_count_output = gr.Number(label="Number of Humans Detected")
        other_objects_output = gr.Textbox(label="Other Objects Detected")
    btn = gr.Button("Detect Objects")

    btn.click(
        process_image_for_gradio,
        inputs=inp_image,
        outputs=[out_image, human_count_output, other_objects_output]
    )

demo.launch(share=True)
```
